```python
from ultralytics import YOLO
import os
import numpy as np
```

```python
ruta = os.getcwd()
rutaModelo = os.path.join(ruta,'..','models','yolov8n.pt')

model = YOLO(rutaModelo)

metrics = model.val(
    data='C:\\Users\\nicoc\\Downloads\\Tesis NNs\\TesisLab\\Deteccion-de-vehiculos-y-estimacion-de-distancias-con-YOLOv8\\data\\val\\data.yaml',
    classes=[2, 3]
)


print(model.names)
print(f"mAP@0.5:      {metrics.box.map50:.3f}")
print(f"mAP@0.5:0.95: {metrics.box.map:.3f}")
print(f"Precision:    {metrics.box.mp:.3f}")
print(f"Recall:       {metrics.box.mr:.3f}")
print(f"F1-score:     {metrics.box.f1}")
print(metrics.box.f1_curve)
print(metrics.box.p_curve)


# Curva F1 de cada clase
f1_car = metrics.box.f1_curve[0]
f1_moto = metrics.box.f1_curve[1]
p_car = metrics.box.p_curve[0]
p_moto = metrics.box.p_curve[1]

# Índice donde cada curva alcanza su máximo
idx_car = np.argmax(f1_car)
idx_moto = np.argmax(f1_moto)
idxp_car = np.argmax(p_car)
idxp_moto = np.argmax(p_moto)


# El umbral en ese índice
umbrales = np.linspace(0, 1, len(f1_car))
umbrales_p = np.linspace(0,1,len(p_car))


print(f"Car  → F1 máx {f1_car[idx_car]:.3f} en umbral {umbrales[idx_car]:.3f}")
print(f"Moto → F1 máx {f1_moto[idx_moto]:.3f} en umbral {umbrales[idx_moto]:.3f}")


print(f"Car  → precision máx {p_car[idxp_car]:.3f} en umbral {umbrales_p[idxp_car]:.3f}")
print(f"Moto → precision máx {p_moto[idxp_moto]:.3f} en umbral {umbrales_p[idxp_moto]:.3f}")
```
